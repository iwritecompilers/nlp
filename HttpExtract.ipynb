```python
import os
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        pass

    def handle_endtag(self, tag):
        pass

    def handle_data(self, data):
        print(data, end=" ")

parser = MyHTMLParser()   
    
    
data_dir = 'data/TREC/trec05p-1/data'
senders = []
dir_limit = 1
for corpus in os.listdir(data_dir):
    file_limit = 100
    for collection in os.listdir(f"{data_dir}/{corpus}/"):
#         print(f"{data_dir}/{corpus}/{collection}")
        fin = open(f"{data_dir}/{corpus}/{collection}", 'rb')
        encoded_message = fin.readlines()
        split_indices = []
#         body_index = 0
        for i, line in enumerate(encoded_message):
#             if len(line.strip()) == 0:
#                 body_index = i
#                 break
            if str.encode("------=_NextPart") in line:
                split_indices.append(i)
            if str.encode("From: ") in line:
                senders.append(line.decode("latin-1"))
#             if str.encode('Return-Path: ') in line:
#                 body_index = i
#                 break
        charset = "latin-1"
        for i in range(len(split_indices) - 1):
            x = split_indices[i]
            y = split_indices[i + 1]
            j = x + 1
            charset_found = False
            while j < y:
                line = encoded_message[j].decode("latin-1").strip()
                if len(line) == 0:
                    break
                if "charset" in line:
                    charset = line.split("=")[1][1:-1]
                j += 1
            print("--- EMAIL: --- ")
            parser.feed(b"".join(encoded_message[j+1:y]).decode(charset))
#         epic = b"".join(encoded_message[body_index+1:]).decode('latin-1', errors='xmlcharrefreplace')
#         print(epic)
#         parser.feed(epic)
        fin.close()
        file_limit -= 1
        if file_limit == 0:
            break
        
    dir_limit -= 1
    if dir_limit == 0:
        break

for sender in senders:
    print(sender)
```
